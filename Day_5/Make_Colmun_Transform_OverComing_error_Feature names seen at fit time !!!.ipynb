### Using column Transformer

```python
ohc = OneHotEncoder(sparse_output=False, drop='first').set_output(transform='pandas')
ct = make_column_transformer((ohc, ['driveway','recroom','fullbase','gashw','airco','prefarea']),
                            ('passthrough',['lotsize','bedrooms','bathrms','stories','garagepl']),verbose_feature_names_out=False)

## price colmun is not inlcuded
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python

ct = ct.set_output(transform='pandas')
x_trans = ct.fit_transform(X)
x_trans.columns
```

```python
## unlabelled data 
tst = pd.read_csv('tst_Housing.csv')
tst_trans = ct.transform(tst)
tst_trans.columns
##  num of  columns in .fit(x,y) and .predict()  must match 
```

```python
# Linear Regression
lr = LinearRegression()
lr.fit(x_trans, y)

y_pred = lr.predict(tst_trans)
y_pred
```

```python
## Using Pipeline
from sklearn.pipeline import Pipeline

pipe = Pipeline([('TRNS', ct),('LR', lr)])
pipe.fit(X,y)
y_pred_pipe = pipe.predict(tst)
y_pred_pipe
```
