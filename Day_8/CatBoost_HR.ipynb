```python
from catboost import CatBoostClassifier

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, KFold
```

```python
df = pd.read_csv('HR_comma_sep.csv', sep=',')
df.head()
```

```python
# dum_hr = pd.get_dummies(df, drop_first=True)
# dum_hr.head(2)
```

```python
X = df.drop('left', axis=1)
y = df['left']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24)
```

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=24)
lgbmc = CatBoostClassifier(random_state=24, cat_features=['Department', 'salary'])

params = {
    'n_estimators': [50,100],
    'learning_rate': [0.01, 0.1, 0.3, 0.5],
    'max_depth': [1,3,5]
}
gcv = GridSearchCV(lgbmc, param_grid=params, cv=kfold, verbose=3, scoring='neg_log_loss')
gcv.fit(X_train, y_train)
print(gcv.best_params_)
print(gcv.best_score_)
```
